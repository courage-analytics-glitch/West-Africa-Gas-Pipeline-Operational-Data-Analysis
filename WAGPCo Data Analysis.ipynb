## Overview

This notebook compiles and analyzes publicly available operational data on the West African Gas Pipeline Company (WAGPCo), operator of the 678 km West African Gas Pipeline connecting Nigeria, Benin, Togo, and Ghana. Built as a records and data management portfolio piece.

## Scope

1. Company profile and ownership structure
2. Pipeline infrastructure specifications
3. Operational performance metrics, gas delivery volumes and asset availability
4. Historical milestones timeline

## Data Sources

Compiled from WAGPCo public statements, company website (wagpco.com), and industry reporting, current as of July 2026. No API exists for this data, records were manually structured from public sources, reflecting the kind of records consolidation this role requires.

## Summary

WAGPCo operates the only cross-border natural gas pipeline in West Africa, a monopoly infrastructure asset rather than a company facing direct competitors. Its real competitive dynamic is against substitute fuels, diesel and heavy fuel oil, which it displaces at lower cost across Ghana, Togo, and Benin.

**Key findings:**
1. Nigeria supplies 68% of total pipeline volume; deliveries grew 153% from 2011 to 2025
2. Ghana is the largest downstream market, saving an estimated $3B versus liquid fuel since 2011, and is now pursuing regional gas hub status
3. WAPCo's tax remittances to member states jumped 156% in 2025, reflecting a strong operational turnaround
4. The Ghana-Cote d'Ivoire expansion, proposed 2017, appears stalled since 2021, while WAGPA has signaled renewed interest in extending to landlocked countries
5. A structural data gap exists in WAGPCo's own public reporting: no source breaks out Ghana, Togo, and Benin's individual shares of downstream gas volume, only Nigeria's aggregate 68% supply share is disclosed

**Note on methodology:** No API or structured dataset exists for this company. All figures were sourced from public statements, news reporting, and regulatory filings, then manually structured into a consistent dataset, reflecting the type of records consolidation this analysis was built to demonstrate.
